```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import scipy as sp
```

```python
# Define axis ranges
x = np.linspace(900, 1000, 500)  # X-axis from 1 to 1000 (log-scale requires positive values)
y = np.linspace(7, 8, 8)   # Y-axis from 0 to 15
X, Y = np.meshgrid(x, y)

# Calculate Z = X^Y
Z_1 = X**Y
Z_2 = sp.special.comb(X, Y, repetition=True)

# Create the plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot with a logarithmic Z-axis
surf_1 = ax.plot_surface(X, Y, Z_1, cmap='viridis')
surf_2 = ax.plot_surface(X, Y, Z_2, cmap='viridis')
ax.set_zscale('log', base=10)  # Set the Z-axis to a logarithmic scale


# Label axes
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis (log scale)')

# Add a color bar for the surface plot
fig.colorbar(surf_2, ax=ax, shrink=0.5, aspect=10)

plt.show()
```

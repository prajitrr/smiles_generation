```python
import pandas as pd
import numpy as np
from rdkit import Chem
```

```python
complete_frame = pd.DataFrame()
file_1 = pd.read_csv("./unjoined_databases/All_reactant_Name_smiles.csv", usecols=['compound_name', 'SMILES'])
file_2 = pd.read_csv("./unjoined_databases/BSH Specificity Calculations 2 - SMILES and Formulas.csv")
file_3 = pd.read_csv("./unjoined_databases/Dorrestein Lab - Chemical Inventory - All(outdated).csv", usecols=['Original Name', 'compound_name', "SMILES_unsalted"])
file_4 = pd.read_csv("./unjoined_databases/Dorrestein Lab - Chemical Inventory - Amines&Other.csv", usecols=['Original name + [notes]', 'compound_name', "SMILES"])
file_5 = pd.read_csv("./unjoined_databases/Dorrestein Lab - Chemical Inventory - BAs.csv", usecols=['Original Name', 'compound_name', "SMILES_unsalted"])
file_6 = pd.read_csv("./unjoined_databases/Dorrestein Lab - Chemical Inventory - MW.csv", skiprows=1, usecols=['Unnamed: 2', 'Name', "SMILES"])
file_7 = pd.read_csv("./unjoined_databases/Dorrestein Lab - Chemical Inventory - OldBAs.csv", usecols=['original_name', 'compound_name', "SMILES_unsalted"])
file_8 = pd.read_csv("./unjoined_databases/Dorrestein Lab - Chemical Inventory - OldFAs.csv", usecols=['compound_name', 'SMILES'])
```

```python
# Process and add in files 1 and 2
file_2 = file_2.dropna(subset=['SMILES'])
file_2["Unnamed: 1"][48] = "CA"
file_2["Unnamed: 1"][89] = "Tau-CA"
whole_names = pd.concat([file_2["Compound Name"], file_2["SMILES"]], axis=1).dropna()
whole_names = whole_names.rename(columns={"Compound Name": "compound_name", "SMILES": "SMILES"})
abbreviations = pd.concat([file_2["Unnamed: 1"], file_2["SMILES"]], axis=1).dropna()
abbreviations = abbreviations.rename(columns={"Unnamed: 1": "compound_name", "SMILES": "SMILES"})

complete_frame = pd.concat([file_1, whole_names, abbreviations], axis=0)


# Process and add in file 3
file_3 = file_3.dropna(subset=['SMILES_unsalted'])
original_names = pd.concat([file_3["Original Name"], file_3["SMILES_unsalted"]], axis=1).dropna()
original_names = original_names.rename(columns={"Original Name": "compound_name", "SMILES_unsalted": "SMILES"})
compound_names = pd.concat([file_3["compound_name"], file_3["SMILES_unsalted"]], axis=1).dropna()
compound_names = compound_names.rename(columns={"compound_name": "compound_name", "SMILES_unsalted": "SMILES"})
complete_frame = pd.concat([complete_frame, original_names, compound_names], axis=0)


# Process and add in file 4

file_4 = file_4.dropna(subset=['SMILES'])
original_names_f4 = pd.concat([file_4["Original name + [notes]"], file_4["SMILES"]], axis=1).dropna()
original_names_f4 = original_names_f4.rename(columns={"Original name + [notes]": "compound_name", "SMILES": "SMILES"})
compound_names_f4 = pd.concat([file_4["compound_name"], file_4["SMILES"]], axis=1).dropna()
compound_names_f4 = compound_names_f4.rename(columns={"compound_name": "compound_name", "SMILES": "SMILES"})
complete_frame = pd.concat([complete_frame, original_names_f4, compound_names_f4], axis=0)


# Process and add in file 5

file_5 = file_5.dropna(subset=['SMILES_unsalted'])
original_names_f5 = pd.concat([file_5["Original Name"], file_5["SMILES_unsalted"]], axis=1).dropna()
original_names_f5 = original_names_f5.rename(columns={"Original Name": "compound_name", "SMILES_unsalted": "SMILES"})
compound_names_f5 = pd.concat([file_5["compound_name"], file_5["SMILES_unsalted"]], axis=1).dropna()
compound_names_f5 = compound_names_f5.rename(columns={"compound_name": "compound_name", "SMILES_unsalted": "SMILES"})
complete_frame = pd.concat([complete_frame, original_names_f5, compound_names_f5], axis=0)


# Process and add in file 6

file_6 = file_6.drop([55,96,118, 136])
file_6 = file_6.dropna(subset=['SMILES'])
unnames_f6 = pd.concat([file_6["Unnamed: 2"], file_6["SMILES"]], axis=1).dropna()
unnames_f6 = unnames_f6[unnames_f6["Unnamed: 2"] != "1"]
unnames_f6 = unnames_f6[unnames_f6["Unnamed: 2"] != "2"]
unnames_f6 = unnames_f6[unnames_f6["Unnamed: 2"] != 1]
unnames_f6 = unnames_f6[unnames_f6["Unnamed: 2"] != 2]
unnames_f6 = unnames_f6.rename(columns={"Unnamed: 2": "compound_name", "SMILES": "SMILES"})
names_f6 = pd.concat([file_6["Name"], file_6["SMILES"]], axis=1).dropna()
names_f6 = names_f6.rename(columns={"Name": "compound_name", "SMILES": "SMILES"})

complete_frame = pd.concat([complete_frame, unnames_f6], axis=0)


# Process and add in file 7

file_7 = file_7.dropna(subset=['SMILES_unsalted'])
original_names_f7 = pd.concat([file_7["original_name"], file_7["SMILES_unsalted"]], axis=1).dropna()
original_names_f7 = original_names_f7.rename(columns={"original_name": "compound_name", "SMILES_unsalted": "SMILES"})
compound_names_f7 = pd.concat([file_7["compound_name"], file_7["SMILES_unsalted"]], axis=1).dropna()
compound_names_f7 = compound_names_f7.rename(columns={"compound_name": "compound_name", "SMILES_unsalted": "SMILES"})

complete_frame = pd.concat([complete_frame, original_names_f7, compound_names_f7], axis=0)


# Process and add in file 8

file_8 = file_8.dropna(subset=['SMILES'])
file_8 = file_8.dropna()

complete_frame = pd.concat([complete_frame, file_8], axis=0)
```

```python
# Validate SMILES in each row

for i, entry in complete_frame.iterrows():
    try:
        entry["SMILES"] = Chem.MolToSmiles(Chem.MolFromSmiles(entry["SMILES"]))
    except:
        entry["SMILES"] = np.nan

complete_frame = complete_frame.dropna(subset=['SMILES'])
```

```python
complete_frame = complete_frame.drop_duplicates()
complete_frame.to_csv("./databases/SMILES_names_master_database.csv", index=False)
```
